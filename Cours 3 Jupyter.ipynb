```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inconue(x):
    return np.cos(10. * x)* np.exp(-x/2)



nb_points = 20
xl, xr = 0,2
esp = 0.1

XS = np.linspace(xl, xr, nb_points)
YS = inconue(XS) + esp * np.random.randn(len(XS))

X = XS.reshape(nb_points,1)

plt.plot(XS, YS, 'o', label='données')
```

```python
modele = LinearRegression(fit_intercept=True)
modele.fit(X, YS)
```

```python
obs = np.linspace(0,2,200)
ordo = modele.predict(obs.reshape(200,1))

plt.plot(obs, ordo, color='red', label='modèle')
```
